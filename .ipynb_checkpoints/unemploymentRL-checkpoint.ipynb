## Q learning with a table

```python
# 1. Load Environment and Q-table structure
env = gym.make('unemployment-v0')
n_states=env.n_states
#print(env.observation_space.shape,env.action_space.n)
Q = np.zeros([n_states,env.action_space.n])
#print(Q.shape)
# env.obeservation.n, env.action_space.n gives number of states and action in env loaded
# 2. Parameters of Q-leanring
eta = .628
gma = 1.0
epis = 1000
epsilon=10.0
rev_list = [] # rewards per episode calculate
# 3. Q-learning Algorithm
for i in range(epis):
    # Reset environment
    #print(i)
    s = env.reset()

    #print('state={state},s={s}'.format(state=state,s=s))
    rAll = 0
    d = False
    j = 0
    #The Q-Table learning algorithm
    while j < 44:
        #env.render()
        j+=1
        # Choose action from Q table
        #print('s={s},Q={Q}'.format(s=s,Q=Q))
        if np.random.uniform(0, 1) < epsilon/(i+1):
            a = np.random.randint(0,2)
        else:
            a = np.argmax(Q[s,:])     
            #if a>1:
            #    print('a={a}'.format(a=a))
            #    print(Q[s,:],s)
        #Get new state & reward from environment
        s1,r,d,_ = env.step(a)

        #Update Q-Table with new knowledge
        Q[s,a] = Q[s,a] + eta*(r + gma*np.max(Q[s1,:]) - Q[s,a])
        rAll += r
        #state = state1
        s=s1
        if d == True:
            break
            
    rev_list.append(rAll)

print("Reward Sum on all episodes " + str(sum(rev_list)/epis))
print("Final Values Q-Table")
print(env.get_empldistributionQ(Q)) #Q)
```

# Policy iteration

```python
def find_optimal_policy():
    env.reset()
    n_states=env.n_states
    n_actions=env.action_space.n
    transitions = np.zeros([n_states,env.action_space.n])
    R = np.zeros([n_states,env.action_space.n])
    Sp = np.zeros([n_states,env.action_space.n]).astype(int)
    pi = np.zeros([n_states]).astype(int)
    V = np.zeros([n_states])
    gamma=0.9
    eps=1e-3

    for s in range(n_states):
        for a in range(n_actions):
            R[s,a],Sp[s,a]=env.get_rewards(s,a)

    #print(R)

    policy_stable=False
    while not policy_stable:
        # Policy evaluation
        theta=0.001
        delta=10.0
        while theta<delta:
            delta=0
            for s in range(n_states):
                v=V[s]
                # deterministic
                if Sp[s,a]<n_states:
                    V[s]=R[s,pi[s]]+gamma*V[Sp[s,pi[s]]]
                else:
                    V[s]=R[s,pi[s]]
                    
                delta=np.maximum(delta,np.abs(V[s]-v))

        # policy improvement
        policy_stable=True
        for s in range(n_states):
            b=pi[s]

            max_Vs=0
            argmax_a=-1
            for a in range(n_actions):
                # deterministic
                if Sp[s,a]<n_states:
                    Vs=R[s,a]+gamma*V[Sp[s,a]]
                else:
                    Vs=R[s,a]
                    
                if Vs>=max_Vs:
                    argmax_a=a

            pi[s]=argmax_a

            if np.abs(b-pi[s])>eps:
                policy_stable=False

    #print(pi,V)
    #print(env.get_empldistribution(pi))
    print(R)
    
    return pi,V
```
