# Kannusteet

```python
import gym
import gym_unemployment
import numpy as np
import matplotlib.pyplot as plt
```

```python
def run_episodes(det=True):
    env = gym.make('unemployment-v0')
    s=env.reset()
    states=[]
    rewards=[]
    totalrewards=np.zeros(100)
    for episode in range(100):
        done=False
        r=0
        i=0
        state=np.zeros((50,5))
        reward=np.zeros(50)

        while not done:
            #env.render()
            if det==1: # unemp
                if s[0]==1:
                    action=1
                else:
                    action=0
                s, rr, done, info = env.step(action)
            elif det==2: # only work
                if s[0]==0:
                    action=1
                else:
                    action=0
                s, rr, done, info = env.step(action)
            elif det==3: # retire early & unemp
                if s[2]>62:
                    action=2
                elif s[0]==1:
                    action=1
                else:
                    action=0
                s, rr, done, info = env.step(action)

            else:
                s, rr, done, info = env.step(env.action_space.sample())
            #print(s,state.shape)
            state[i,:]=np.array(s)
            reward[i]=rr
            r=r+reward[i]
            i=i+1
            if done:
                break

        states.append(state)
        rewards.append(reward)
        totalrewards[episode]=r

        s=env.reset()
    return states,rewards,totalrewards

def plot_episodes(states,rewards,totalrewards):
    fig, axs = plt.subplots(6, 1)
    for k in range(100):
        s=states.pop()
        axs[0].plot(s[:,0])
        axs[1].plot(s[:,1])
        axs[2].plot(s[:,2])
        axs[3].plot(s[:,3])
        axs[4].plot(s[:,4])
        r=rewards.pop()
        axs[5].plot(r)

    axs[5].set_xlabel('time')
    axs[0].set_ylabel('states')
    axs[1].set_ylabel('prev_unemp')
    axs[2].set_ylabel('age')
    axs[3].set_ylabel('wage')
    axs[4].set_ylabel('pension')
    axs[5].set_ylabel('reward')

    for k in range(6):
        axs[k].grid(True)
        axs[k].set_xlim(0, 50)

    plt.show()
    plt.hist(totalrewards)



```

```python
states,rewards,totalrewards=run_episodes(1)
plot_episodes(states,rewards,totalrewards)            
```

```python
# do not work
states,rewards,totalrewards=run_episodes(0)
plot_episodes(states,rewards,totalrewards)            

```

```python
# work
states,rewards,totalrewards=run_episodes(2)
plot_episodes(states,rewards,totalrewards) 
```

```python
# retire early
states,rewards,totalrewards=run_episodes(3)
plot_episodes(states,rewards,totalrewards) 
```
