```python
#Import required libraries
import sys
from PIL import Image, ImageDraw
import math
import numpy as np
```

```python
#Create Image object
size = (500,500)
im = Image.new('RGBA', size, "red")

def rotate(x0,a,theta):
    b=np.zeros(2)
    c=np.zeros(2)
    b[0]=a[0]-x0[0]
    b[1]=a[1]-x0[1]
    c[0]=b[0]*math.cos(theta)-b[1]*math.sin(theta)+x0[0]
    c[1]=b[1]*math.cos(theta)+b[0]*math.sin(theta)+x0[1]
    
    return c

def draw_line(draw,x0,a1,a2,theta):
    c1=rotate(x0,a1,theta)
    c2=rotate(x0,a2,theta)
    
    draw.line((c1[0], c1[1], c2[0], c2[1]), fill=(0, 0, 0), width=40)

def draw_swastika(im,shift=0,theta=0):
    #Draw line swatika
    draw = ImageDraw.Draw(im)
    draw.ellipse([100,100,400,400],fill="white")
    draw.line((350, 170, 230-shift, 170), fill=(0, 0, 0), width=40)
    draw.line((250-shift, 170, 250+shift, 350), fill=(0, 0, 0), width=40)
    draw.line((150, 330, 250+shift, 330), fill=(0, 0, 0), width=40)

    if shift<1:
        x0=np.array([250,250])
        a1=np.array([170,150])
        a2=np.array([170,250])
        draw_line(draw,x0,a1,a2,theta)
        a1=np.array([150,250])
        a2=np.array([350,250])
        draw_line(draw,x0,a1,a2,theta)
        a1=np.array([330,250])
        a2=np.array([330,350])
        draw_line(draw,x0,a1,a2,theta)
        #draw.line((170, 150, 170, 250), fill=(0, 0, 0), width=40)
        #draw.line((150, 250, 350, 250), fill=(0, 0, 0), width=40)
        #draw.line((330, 250, 330, 350), fill=(0, 0, 0), width=40)

def draw_z(im):
    #Draw line swatika
    draw = ImageDraw.Draw(im)
    draw.ellipse([100,100,400,400],fill="white")
    draw.line((150, 170, 350, 170), fill=(0, 0, 0), width=40)
    draw.line((350, 170, 150, 350), fill=(0, 0, 0), width=40)
    draw.line((150, 350, 350, 350), fill=(0, 0, 0), width=40)
    
images = []

for k in range(10):
    draw_swastika(im,theta=0,shift=0)
    images.append(im)

for theta in np.arange(0,math.pi/2,math.pi/2/50):
    print(theta)
    im = Image.new('RGB', size, "red")
    draw_swastika(im,theta=theta,shift=0)
    images.append(im)

for k in range(3):
    draw_swastika(im,theta=theta,shift=0)
    images.append(im)
    
    
for shift in np.arange(1,80,1):
    print(shift)
    im = Image.new('RGB', size, "red")
    draw_swastika(im,shift=shift)
    images.append(im)

images[0].save('pillow_imagedraw.gif',
               save_all=True, append_images=images[1:], optimize=False, duration=40, loop=0)
    
    
#draw_z(im)


#Show image
im.show()
```
